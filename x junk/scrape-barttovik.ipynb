```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# https://barttorvik.com/playerstat.php?link=y&minGP=1&minpick=60&year=all&minmin=5&start=20251101&end=20260501&pickSelect=-1

# Target URL (e.g., pulling all individual player stats tables)
url = "https://barttorvik.com"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

print("Fetching webpage...")
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the main data table in the DOM
table = soup.find('table') 

player_rows = []
if table:
    # Target table body rows
    rows = table.find_all('tr')
    
    for row in rows:
        # Extract table data fields
        cols = row.find_all(['td', 'th'])
        cols_text = [ele.text.strip() for ele in cols]
        
        # Filter out spacer rows or section dividers
        if len(cols_text) > 5 and cols_text[0] != 'Rk':
            player_rows.append(cols_text)

# Build out the DataFrame manually
if player_rows:
    df_scraped = pd.DataFrame(player_rows)
    df_scraped.to_csv("barttorvik_scraped_players.csv", index=False)
    print(f"Successfully scraped {len(df_scraped)} players via HTML soup!")
else:
    print("Table parsing failed. The page might be rendering components via JavaScript.")
```
